```python
import matplotlib.pyplot as plt
import numpy as np

# x  
x = np.arange(-1, 1.05, 0.05)

# xx   
xx_values = np.arange(-1, 1.5, 0.5)

#   ( xx    )
colors = plt.cm.viridis(np.linspace(0, 1, len(xx_values)))

#  
for i, xx in enumerate(xx_values):
    y = x**2 - 2 * 0.5 * (x - xx) * np.sign(x - xx)
    plt.plot(x, y, label=f'xx={xx:.2f}', color=colors[i])

#   
plt.xlabel('x')
plt.ylabel('y')
plt.title('Potential Evolution with xx')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# x  
x = np.arange(-1.5, 1.55, 0.05)

# xx   
xx_values = np.arange(-1, 1.3, 0.5)

#  
for xx in xx_values:
    #   
    plt.figure()
    
    # y  
    y = x**2 - 2 * 0.5 * (x - xx) * np.sign(x - xx)
    
    #  
    plt.plot(x, y, label=f'xx={xx:.2f}')
    
    #   
    plt.xlabel('X')
    plt.ylabel('Potential')
   
    plt.legend()
   
    
    #   ( )
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-1, 1.01, 0.01)

plt.figure(figsize=(10,8))

xx=0.
y = x**2 - 2 * 0.5 * (x - xx) * np.sign(x - xx)
plt.plot(x, y, label=f'A', color='r')


xx=-.5
y = x**2 - 2 * 0.5 * (x - xx) * np.sign(x - xx)


plt.plot(x, y, label=f'B', color='b')


plt.xlabel('X', fontsize=30)
plt.ylabel('Potential',fontsize=30)
plt.xticks([])
plt.yticks([])


plt.legend(fontsize = 20)
plt.gca().set_xticklabels([])
plt.gca().set_yticklabels([]) 
plt.scatter(-0.5, 0.28, color='g', s = 200)
plt.show()
```
